## Data Scraping from Fbref

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# For premier League
base_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
```

```python
web_data = requests.get(base_url)
web_data
```

```python
soup = BeautifulSoup(web_data.text)
```

```python
stats_table = soup.select('table.stats_table')
stats_table
```
